## Analitik

Penentuan lokasi ini menggunakan teknologi kecerdasan buatan (AI) yang mana akan melakukan clustering untuk menentukan lokasi penempatan pusat data. Analisis klaster atau clustering adalah tugas untuk mengelompokkan sekumpulan objek sedemikian rupa sehingga objek-objek dalam kelompok yang sama (disebut cluster) lebih mirip (dalam beberapa hal) satu sama lain daripada objek-objek dalam kelompok lain (cluster). Ini adalah tugas utama analisis data eksplorasi, dan teknik umum untuk analisis data statistik, yang digunakan di banyak bidang, termasuk pengenalan pola, analisis gambar, pengambilan informasi, bioinformatika, kompresi data, grafik komputer, dan pembelajaran mesin.

Analisis klaster itu sendiri bukanlah satu algoritma tertentu, tetapi tugas umum yang harus diselesaikan. Hal ini dapat dicapai dengan berbagai algoritma yang berbeda secara signifikan dalam pemahaman mereka tentang apa yang merupakan cluster dan bagaimana menemukan mereka secara efisien. Pengertian populer dari cluster termasuk grup dengan jarak kecil antara anggota cluster, area padat dari ruang data, interval atau distribusi statistik tertentu. Oleh karena itu, pengelompokan dapat dirumuskan sebagai masalah optimasi multi-tujuan. Algoritma pengelompokan yang sesuai dan pengaturan parameter (termasuk parameter seperti fungsi jarak yang digunakan, ambang kepadatan atau jumlah klaster yang diharapkan) bergantung pada kumpulan data individual dan tujuan penggunaan hasil. Analisis klaster seperti itu bukanlah tugas otomatis, tetapi proses berulang dari penemuan pengetahuan atau optimasi multi-tujuan interaktif yang melibatkan percobaan dan kegagalan. Hal ini sering diperlukan untuk memodifikasi data preprocessing dan parameter model sampai hasilnya mencapai properti yang diinginkan.

Tipe algoritma clustering yang akan digunakan adalah clustering berdasarkan kepadatan. Dalam pengelompokan berbasis kepadatan, cluster didefinisikan sebagai area dengan kepadatan lebih tinggi daripada sisa kumpulan data. Objek di area yang jarang - yang diperlukan untuk memisahkan cluster - biasanya dianggap sebagai titik kebisingan dan batas.

Metode pengelompokan berbasis kepadatan yang paling populer adalah DBSCAN. Berbeda dengan banyak metode yang lebih baru, ia menampilkan model cluster yang terdefinisi dengan baik yang disebut "density-reachability". Mirip dengan pengelompokan berbasis tautan, ini didasarkan pada titik penghubung dalam ambang jarak tertentu. Namun, itu hanya menghubungkan titik-titik yang memenuhi kriteria kepadatan, dalam varian asli yang didefinisikan sebagai jumlah minimum objek lain dalam radius ini. Sebuah cluster terdiri dari semua objek yang terhubung dengan kepadatan (yang dapat membentuk sebuah cluster dengan bentuk yang berubah-ubah, berbeda dengan banyak metode lain) ditambah semua objek yang berada dalam jangkauan objek ini. Properti lain yang menarik dari DBSCAN adalah bahwa kompleksitasnya cukup rendah ‚Äì memerlukan sejumlah linear query range pada database ‚Äì dan pada dasarnya akan menemukan hasil yang sama (ini deterministik untuk titik inti dan noise, tetapi tidak untuk titik perbatasan) di setiap run, oleh karena itu tidak perlu menjalankannya berkali-kali. OPTICS adalah generalisasi DBSCAN yang menghilangkan kebutuhan untuk memilih nilai yang sesuai untuk parameter rentang ùú∫, dan menghasilkan hasil hierarki yang terkait dengan pengelompokan tautan. DeLi-Clu, Density-Link-clustering menggabungkan ide-ide dari single-linkage clustering dan OPTICS, menghilangkan parameter ùú∫ sepenuhnya dan menawarkan peningkatan kinerja melalui OPTICS dengan menggunakan indeks R-tree.

Kelemahan utama DBSCAN dan OPTICS adalah mereka mengharapkan semacam penurunan kepadatan untuk mendeteksi batas cluster. Pada kumpulan data dengan, misalnya, distribusi Gaussian yang tumpang tindih - kasus penggunaan umum dalam data buatan - batas cluster yang dihasilkan oleh algoritma ini akan sering terlihat sewenang-wenang, karena kepadatan cluster menurun terus menerus. Pada kumpulan data yang terdiri dari campuran Gauss, algoritma ini hampir selalu mengungguli metode seperti pengelompokan EM yang mampu memodelkan jenis data ini dengan tepat.

# Modelling

## Region: Sumbawa

```python
df_sumbawa_info = df_region[df_region["region"] == "Sumbawa Region, Indonesia"]
df_sumbawa = df_sumbawa_info.iloc[:,[0,1,2]]
df_sumbawa.head()
```

```python
df_sumbawa.info()
```

### Perhiutngan _Elbow Point Graph_
_Elbow Point Graph_ adalah salah satu metode yang populer dalam menentukan jumlah optimal k pada teknik _clustering_ K-Means. Apabila grafik membentuk sebuah lengan, maka bagian "siku" menandakan indikasi bahwa model tersebut fit untuk digunakan

```python
cluster_sumbawa = df_sumbawa.iloc[:,[0,1]].values
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters =i, init='k-means++', random_state=42)
  kmeans.fit(cluster_sumbawa)

  wcss.append(kmeans.inertia_)

fig = plt.figure()
ax = fig.add_subplot(111, xlabel="Number of Cluster Sumbawa", ylabel="WCSS Sumbawa")
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph Sumbawa')

plt.show()
#show the optimal number of clusters
```

```python
#the optimal number is 2 (shown on the figure above, when on cluster 2 the wcss value significantly change too!)
kmeans = KMeans(n_clusters= 2, init='k-means++', random_state=42)
Y_sumbawa = kmeans.fit_predict(df_sumbawa)
print(np.unique(Y_sumbawa))
```

### Visualisasi Posisi Gempa Sumbawa

```python
plt.figure(figsize=(10,8))
plt.scatter(cluster_sumbawa[Y_sumbawa==0,0], cluster_sumbawa[Y_sumbawa==0,1], s=60, c='green', label='cluster_sumbawa 1')
plt.scatter(cluster_sumbawa[Y_sumbawa==1,0], cluster_sumbawa[Y_sumbawa==1,1], s=60, c='red', label='cluster_sumbawa 2')
# plt.scatter(cluster_sumbawa[Y==2,0], cluster_sumbawa[Y==2,1], s=60, c='lime', label='cluster_sumbawa 3')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='black')

plt.title('posisi gempa Sumbawa')
plt.xlabel('lintang')
plt.ylabel('bujur')
plt.show()
```
